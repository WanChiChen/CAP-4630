#HW2 Problem 1


Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

    p1 is a minimum of f1
    p2 is a maximum of f2
    p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots.

#f1

Let f1 be equal to:
$$ f_1 = 0.5(x + y) $$

The local minimum of f1 is at p1 = (0,0,0)

#f2

Let f2 be equal to:
$$ f_2 = x*y $$

The local maximum of f2 is at p2 = (10,10,100)

##3D Plot

```
def f2(x,y):
  return x*y

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

yline = np.linspace(0, 10, 100)
xline = np.linspace(0, 10, 100)
X, Y = np.meshgrid(xline, yline)
Z = f2(X,Y)

ax.plot_surface(X,Y,Z)
ax.scatter(10,10,100, c='Red')
plt.show()
```

##Density Plot

```
plt.contourf(X,Y,Z, cmap='RdGy')
plt.colorbar()
plt.scatter(10,10)
plt.show()
```

##Contour Plot

```
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, colors='black')
plt.colorbar()
plt.scatter(10,10)
plt.show()
```

#f3

Let f3 be equal to:

$$ f_3 = x * cos(y) $$

$$ \nabla f_3(0,0) = ( 0 * sin(0), cos(0) ) = (0,0) $$

The gradient at $ p3 = (0,0,0) $ is the zero vector.

##3D Plot

```
def f2(x,y):
  return x*np.cos(y)
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

yline = np.linspace(-10, 10, 100)
xline = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(xline, yline)
Z = f2(X,Y)

ax.plot_surface(X,Y,Z)
ax.scatter( 0,0,0, c='Red')
ax.view_init(40, 0)
plt.show()
```

##Density Plot

```
plt.contourf(X,Y,Z, cmap='RdGy')
plt.colorbar()
plt.scatter(0,0)
plt.show()
```

##Contour Plot

```
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, colors='black')
plt.scatter(0,0)
plt.show()
```
