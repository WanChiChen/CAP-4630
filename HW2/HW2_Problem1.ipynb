Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

    p1 is a minimum of f1
    p2 is a maximum of f2
    p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots.

#f1

##3D Plot

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
  return 0.5*(x+y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

yline = np.linspace(0, 10, 100)
xline = np.linspace(0, 10, 100)
X, Y = np.meshgrid(xline, yline)
zline = f1(X,Y)

ax.contour3D(X, Y, Z, 50, cmap='binary')

plt.scatter(0,0,25)
plt.show()
```

##Density Plot

#f2

##3D Plot

```
def f2(x,y):
  return x*y

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

yline = np.linspace(0, 10, 100)
xline = np.linspace(0, 10, 100)
X, Y = np.meshgrid(xline, yline)
Z = f2(X,Y)

ax.plot_surface(X,Y,Z)
ax.scatter(10,10,100, c='Red')
plt.show()
```

##Density Plot

```
plt.contourf(X,Y,Z, cmap='RdGy')
plt.colorbar()
plt.scatter(10,10)
plt.show()
```

##Contour Plot

```
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, colors='black')
plt.colorbar()
plt.scatter(10,10)
plt.show()
```

#f3

##3D Plot

```
def f2(x,y):
  return (x-x)+5
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

yline = np.linspace(0, 10, 100)
xline = np.linspace(0, 10, 100)
X, Y = np.meshgrid(xline, yline)
Z = f2(X,Y)

ax.plot_surface(X,Y,Z)
ax.scatter(10,10,5, c='Red')
ax.view_init(5, 10)
plt.show()
```

##Density Plot

```
plt.contourf(X,Y,Z, cmap='RdGy')
plt.colorbar()
plt.scatter(10,10)
plt.show()
```

##Contour Plot

```
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, colors='black')
plt.scatter(10,10)
plt.show()
```
